### Autoencoder -- Image compressor

The encoder will consist in a stack of Conv2D and MaxPooling2D layers (max pooling being used for spatial down-sampling), while the decoder will consist in a stack of Conv2D and UpSampling2D layers.

```python
from keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D, Flatten
from keras.models import Model
import tensorflow.keras as keras
from keras.callbacks import EarlyStopping


list_clothes = outfits_df.columns[2:]
weights = []

for clothing in list_clothes:
    
    #Creating the encoder and decoder
    input_layer = Input(shape=(64, 64, 3)) # there are 3 channels

    # Encoder / compressing
    conv1 = Conv2D(496, (3,3), activation='relu', padding='same')(input_layer)
    pool1 = MaxPooling2D(pool_size=(2,2))(conv1)

    conv2 = Conv2D(248, (3,3), activation='relu', padding='same')(pool1)
    pool2 = MaxPooling2D(pool_size=(2,2))(conv2)

    conv3 = Conv2D(124, (3,3), activation='relu', padding='same')(pool2)
    pool3 = MaxPooling2D(pool_size=(2,2))(conv3)

    conv4 = Conv2D(64, (3,3), activation='relu', padding='same')(pool3)
    pool4 = MaxPooling2D(pool_size=(2,2))(conv4)

    conv5 = Conv2D(32, (3,3), activation='relu', padding='same')(pool4)
    pool5 = MaxPooling2D((2,2), padding='same')(conv5)

    conv6 = Conv2D(16, (3,3), activation='relu', padding='same')(pool5)
    pool6 = MaxPooling2D((2,2), padding='same')(conv6)

    conv7 = Conv2D(16, (3,3), activation='relu', padding='same')(pool6)
    pool7 = MaxPooling2D((2,2), padding='same')(conv7)

    encoded = Flatten(name = 'encoded')(pool7) # Layer embedded

    ## Decoder / uncompress, returning the original form        
    conv8 = Conv2D(16, (3,3), activation='relu', padding='same')(pool7)
    up1 = UpSampling2D((2,2))(conv8)

    conv9 = Conv2D(16, (3,3), activation='relu', padding='same')(up1)
    up2 = UpSampling2D((2,2))(conv9)

    conv10 = Conv2D(32, (3,3), activation='relu', padding='same')(up2)
    up3 = UpSampling2D((2,2))(conv10)

    conv11 = Conv2D(64, (3,3), activation='relu', padding='same')(up3)
    up4 = UpSampling2D((2,2))(conv11)

    conv12 = Conv2D(128, (3,3), activation='relu', padding='same')(up4)
    up5 = UpSampling2D((2,2))(conv12)

    conv13 = Conv2D(248, (3,3), activation='relu', padding='same')(up5)
    up6 = UpSampling2D((1,1))(conv13)

    conv14 = Conv2D(496, (3,3), activation='relu', padding='same')(up6)
    up7 = UpSampling2D((2,2))(conv14)

    decoded = Conv2D(3,(1,1), activation='sigmoid', padding='same')(up7)

    # Creating the autoencoder
    autoencoder = Model(input_layer, decoded)
    autoencoder.compile(optimizer='adadelta', loss='binary_crossentropy')
    encoder = Model(input_layer, encoded)

    # Fit the model created
    features = []
    for i in range(len(outfits_df[clothing])):
        if not pd.isna(outfits_df[clothing][i]):
            features.append(prepare_images(outfits_df[clothing][i][0]))
        else:
            nan_array = np.empty((64, 64, 3))
            nan_array.fill(0)
            features.append(nan_array)
    
    # Convert the list of images into a single NumPy array
    x_train = np.array(features)
    print("Tamano del conjunto train:",len(x_train))
    early_stopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=10, verbose=1, mode='auto')
    autoencoder.fit(x_train, x_train, epochs=20, batch_size=100, validation_split=.15, callbacks=[early_stopping])

    # Savings the weights
    weights.append(encoder.get_weights())
    
    
```

```python
from keras.layers import Input, Dense
from keras.models import Model, Sequential
from tensorflow.keras.layers import concatenate
import numpy as np
from keras import Input, Model

# Creatin the model
# def build_model(input_shapes, n_target_classes=None):
# The categories
categories = outfits_df.columns[2:]

model = Sequential()

# Creating inputs
input_21 = Input(shape=input_shapes['21'], name='21')
input_25 = Input(shape=input_shapes['25'], name='25')
input_237 = Input(shape=input_shapes['237'], name='237')
input_49 = Input(shape=input_shapes['49'], name='49')
input_17 = Input(shape=input_shapes['17'], name='17')
input_27 = Input(shape=input_shapes['27'], name='27')
input_19 = Input(shape=input_shapes['19'], name='19')
input_11 = Input(shape=input_shapes['11'], name='11')
input_104 = Input(shape=input_shapes['104'], name='104')
input_24 = Input(shape=input_shapes['24'], name='24')
input_29 = Input(shape=input_shapes['29'], name='29')
input_41 = Input(shape=input_shapes['41'], name='41')
input_241 = Input(shape=input_shapes['241'], name='241')
input_240 = Input(shape=input_shapes['240'], name='240')
input_258 = Input(shape=input_shapes['258'], name='258')
input_18 = Input(shape=input_shapes['18'], name='18')


# Fully connected layer
layer_21 = Dense(32, activation='relu', name='output_21')(input_21)
layer_25 = Dense(32, activation='relu', name='output_25')(input_25)
layer_237 = Dense(32, activation='relu', name='output_237')(input_237)
layer_49 = Dense(32, activation='relu', name='output_49')(input_49)
layer_17 = Dense(32, activation='relu', name='output_17')(input_17)
layer_27 = Dense(32, activation='relu', name='output_27')(input_27)
layer_19 = Dense(32, activation='relu', name='output_19')(input_19)
layer_11 = Dense(32, activation='relu', name='output_11')(input_11)
layer_104 = Dense(32, activation='relu', name='output_104')(input_104)
layer_24 = Dense(32, activation='relu', name='output_24')(input_24)
layer_29 = Dense(32, activation='relu', name='output_29')(input_29)
layer_41 = Dense(32, activation='relu', name='output_41')(input_41)
layer_241 = Dense(32, activation='relu', name='output_241')(input_241)
layer_240 = Dense(32, activation='relu', name='output_240')(input_240)
layer_258 = Dense(32, activation='relu', name='output_258')(input_258)
layer_18 = Dense(32, activation='relu', name='output_18')(input_18)

"""
Se obtienen las segundas capas las cuales van a ser usadas para crear la primera
capa de completamente conectada. Posteriomrente los resultados se van a concatenar
"""

layer2_21 = Dense(32, activation='relu')(layer_21)
layer2_21 = Model(inputs=input_21, outputs=layer2_21)

layer2_25 = Dense(32, activation='relu')(layer_25)
layer2_25 = Model(inputs=input_25, outputs=layer2_25)

layer2_237 = Dense(32, activation='relu')(layer_237)
layer2_237 = Model(inputs=input_237, outputs=layer2_237)

layer2_49 = Dense(32, activation='relu')(layer_49)
layer2_49 = Model(inputs=input_49, outputs=layer2_49)

layer2_17 = Dense(32, activation='relu')(layer_17)
layer2_17 = Model(inputs=input_17, outputs=layer2_17)

layer2_27 = Dense(32, activation='relu')(layer_27)
layer2_27 = Model(inputs=input_27, outputs=layer2_27)

layer2_19 = Dense(32, activation='relu')(layer_19)
layer2_19 = Model(inputs=input_19, outputs=layer2_19)

layer2_11 = Dense(32, activation='relu')(layer_11)
layer2_11 = Model(inputs=input_11, outputs=layer2_11)

layer2_104 = Dense(32, activation='relu')(layer_104)
layer2_104 = Model(inputs=input_104, outputs=layer2_104)

layer2_24 = Dense(32, activation='relu')(layer_24)
layer2_24 = Model(inputs=input_24, outputs=layer2_24)

layer2_29 = Dense(32, activation='relu')(layer_29)
layer2_29 = Model(inputs=input_29, outputs=layer2_29)

layer2_41 = Dense(32, activation='relu')(layer_41)
layer2_41 = Model(inputs=input_41, outputs=layer2_41)

layer2_241 = Dense(32, activation='relu')(layer_241)
layer2_241 = Model(inputs=input_241, outputs=layer2_241)

layer2_240 = Dense(32, activation='relu')(layer_240)
layer2_240 = Model(inputs=input_240, outputs=layer2_240)

layer2_258 = Dense(32, activation='relu')(layer_258)
layer2_258 = Model(inputs=input_258, outputs=layer2_258)

layer2_18 = Dense(32, activation='relu')(layer_18)
layer2_18 = Model(inputs=input_18, outputs=layer2_18)

# Concatenando las layers
layers_concatenated = concatenate([
    layer2_21.output,
    layer2_25.output,
    layer2_237.output,
    layer2_49.output,
    layer2_17.output,
    layer2_27.output,
    layer2_19.output,
    layer2_11.output,
    layer2_104.output,
    layer2_24.output,
    layer2_29.output,
    layer2_41.output,
    layer2_241.output,
    layer2_240.output,
    layer2_258.output,
    layer2_18.output
])

z = Dense(4, activation='relu')(layers_concatenated)
z = Dense(1, activation='linear')(z)

model = Model(inputs=[
    layer2_21.input,
    layer2_25.input,
    layer2_237.input,
    layer2_49.input,
    layer2_17.input,
    layer2_27.input,
    layer2_19.input,
    layer2_11.input,
    layer2_104.input,
    layer2_24.input,
    layer2_29.input,
    layer2_41.input,
    layer2_241.input,
    layer2_240.input,
    layer2_258.input,
    layer2_18.input
], outputs=z)
```
